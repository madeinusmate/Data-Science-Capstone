```python
#installing libraries

!pip install lxml html5lib beautifulsoup4
import pandas as pd
```

```python
#scrape the page
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df=pd.read_html(url)

#extract only the first table on the page
df=df[0]
df
```

```python
#removing Not assigned Borough

df=df[df.Borough !='Not assigned']
df
```

```python
#reindexing the dataframe

df=df.reset_index()

df=df.drop(["index"], axis=1)

df
```

```python
#looking for Not assigned Neighborhood
print("There are ",df.loc[df.Neighborhood == 'Not assigned', 'Neighborhood'].count(),"Neighborhood with no value assigned")
```

```python
print("the shape of the dataframe is ", df.shape)
```

```python
#importing libraries
!pip install pgeocode
import pgeocode
```

```python
#extract an array of Zipcodes from the dataframe
postal_code=df["Postal Code"].values
```

```python
#get the coordinates for all zipcodes
nomi = pgeocode.Nominatim('ca')
geocoord= nomi.query_postal_code(postal_code)
geocoord
```

```python
#add coordinates column to dataframe

df["Latitude"]=geocoord["latitude"]
df["Longitude"]=geocoord["longitude"]
df
```
